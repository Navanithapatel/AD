```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df=read_csv('/content/pima-indians-diabetes.data.csv')
df.head()
```

```python
file_name='/content/pima-indians-diabetes.data.csv'
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(file_name,names=names)
dataframe.head()
```

```python
array=dataframe.values
X=array[:,0:8]
Y=array[:,8]
```

```python
num_trees=100
seed=7
kfold=KFold(n_splits=10,random_state=seed,shuffle=True)
model=RandomForestClassifier(n_estimators=num_trees,max_features=3)
results=cross_val_score(model,X,Y,cv=kfold)
kfold
```

```python
result=cross_val_score(model,X,Y,cv=kfold)
print(result.mean())
```
