```python
import seaborn as sns
import matplotlib.pyplot as plt
iris_data=sns.load_dataset('iris')
iris_data
```

```python
sns.boxplot(x=iris_data['sepal_width'],palette=['violet','red','lightgreen'],y=iris_data['species'])
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('taxis')
df
```

```python
sns.scatterplot(x=df['fare'],y=df['total'],palette=['yellow','red'])
plt.show()
```

```python
sns.scatterplot(x=df['fare'],y=df['total'],palette=['green','red'],hue=df['payment'])
plt.show()
```

```python
sns.stripplot(x=df['fare'],y=df['payment'])
```

```python
sns.stripplot(x=df['fare'],y=df['payment'],palette=['yellow','red'])
```

```python
sns.stripplot(x=df['fare'],y=df['payment'],palette=['lightgreen','blue'],hue=df['color'])
```

```python
sns.swarmplot(x=df['fare'],y=df['payment'],palette=['blue','red'],hue=df['color'])
```

```python
sns.lineplot(x=df['total'],y=df['fare'])
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=df['payment'],palette=['green','purple'])
```

```python
sns.countplot(x=df['payment'])
```
